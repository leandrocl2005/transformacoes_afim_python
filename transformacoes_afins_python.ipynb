```python
# vértices
vertices = np.array([
    [-1, 1,  1, -1, -1], # x
    [ 1, 1, -1, -1,  1], # y
    [ 1, 1,  1,  1,  1]  # z=1
])
```

```python
# transformação Afim
M1 = np.array([
    [1, 0, 0],
    [0, 2, 5]
])

# transformação Afim
M2 = np.array([
    [2, 0, 5],
    [0, 1, 0]
])

# transformação Afim
M3 = np.array([
    [np.cos(np.pi/4), -np.sin(np.pi/4), 5],
    [np.sin(np.pi/4),  np.cos(np.pi/4), 5]
])
```

```python
# aplicando as transformações
x, y, z = vertices
new_x1, new_y1 = np.matmul(M1, vertices)
new_x2, new_y2 = np.matmul(M2, vertices)
new_x3, new_y3 = np.matmul(M3, vertices)
```

```python
# mostrando os resultados
plt.plot(x, y, color="blue") # original em azul
plt.plot(new_x1, new_y1, color="red")
plt.plot(new_x2, new_y2, color="green")
plt.plot(new_x3, new_y3, color="yellow")
plt.axis('equal')
plt.show()
```

```python
# lendo a imagem
image = cv2.imread("quadro.jpg")
# mudando a imagem para rgb
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
# definindo a transformação
M = np.float32([
    [np.cos(np.pi/4), -np.sin(np.pi/4), 250],
    [np.sin(np.pi/4),  np.cos(np.pi/4),-150]])
# guardando as dimensões da imagem
linhas, colunas, camadas = image.shape
# aplicando a transformação
image = cv2.warpAffine(image, M, (colunas, linhas))
# mostrando imagem
im = plt.imshow(image, interpolation="none")
plt.axis("off")
plt.show()
```

```python
# lendo a imagem
image = cv2.imread("quadro.jpg")

# função para pegar clicks
def click_and_print(event, x, y, flags, param):
    global clicks
    if event == cv2.EVENT_LBUTTONDOWN:
        clicks.append([x,y])
    return 0

# exibindo a janela para pegar os clicks
clicks = []
cv2.namedWindow("image")
test = cv2.setMouseCallback("image", click_and_print)
cv2.imshow("image", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
print(clicks)
```
